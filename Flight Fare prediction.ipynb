### Shapiro-Wilk Test

**Shapiro-Wilk test** is a statistical test that evaluates the null hypothesis:

**H (Null Hypothesis):** The data is normally distributed.

**H (Alternative Hypothesis):** The data is not normally distributed.

```python
# Testing the normality of the price column using Shapiro-Wilk test
stats.shapiro(combined_df['price'])
```

The p-value of `price` is 0.0 which is < 0.05, we are going to reject the null hypothesis, indicating that `price` is **not normally distributed**.
The statistic (0.75) is far from 1, confirming a significant deviation from normality.

### Why Perform Kruskal-Wallis on the Additional Groups?

Since the **Shapiro-Wilk** test already confirmed that `price` is not **normally distributed**, the **Kruskal-Wallis test (a non-parametric test)** is appropriate for checking whether multiple categorical groups affect the **target variable (price)**.

### Performing Kruskal Wallis Statistical Test for Categorical Variables

**Kruskal Wallis Statistical Test**: Comparing differences between two or more independent groups on a continuous or ordinal dependent variable

```python
from scipy import stats

# List of categorical columns
#categorical_cols = ['airline', 'from', 'Class', 'stop', 'Week', 'day_of_week', 'departure_time']
categorical_cols = ['airline', 'from', 'Class', 'stop', 'to', 'Week', 'day_of_week', 'departure_time', 'arrival_time']

def kruskal_wallis_test(df, categorical_col, target_col):
    """Perform Kruskal-Wallis test for a given categorical column against the target variable."""
    
    groups = [df[df[categorical_col] == category][target_col] for category in df[categorical_col].unique()]
    stat, p_value = stats.kruskal(*groups)
    
    print(f"Kruskal-Wallis Test for {categorical_col}: Statistic={stat}, p-value={p_value}")

# Perform Kruskal-Wallis test for all categorical variables
for col in categorical_cols:
    kruskal_wallis_test(combined_df, col, 'price')
```

**Main Takeaways from the Kruskal-Wallis Test Results:**

- All categorical variables significantly impact price (p-value  0.0), meaning these features cause notable differences in flight prices.

- Most influential factors:

    - Class (192,889.58): Strongest impact, likely due to the price difference between business and economy classes.

    - Airline (94,731.64): Major influence, reflecting different pricing strategies.

    - Week (27,602.46): Significant impact due to seasonal trends and demand fluctuations.

    - Stop (20,537.55): More stops typically result in lower prices.

- Moderate impact features:

    - Arrival and departure time (6,532.88 & 2,570.29): Time of travel influences price, possibly due to peak vs. off-peak hours.

# Now Creating generalized functions to calculate various metrics for models:

Create a generalized function to calculate the RMSE for train and test set.

```python
def get_train_rmse(model):
    train_pred = model.predict(X_train)
    mse_train = mean_squared_error(y_train, train_pred)
    rmse_train = round(np.sqrt(mse_train), 4)
    return(rmse_train)
```

```python
def get_test_rmse(model):
    test_pred = model.predict(X_test)
    mse_test = mean_squared_error(y_test, test_pred)
    rmse_test = round(np.sqrt(mse_test), 4)
    return(rmse_test)
```

Create a generalized function to calculate the MAPE for test set.

```python
def mape(actual, predicted):
    return (np.mean(np.abs((actual - predicted) / actual)) * 100)

def get_test_mape(model):
    test_pred = model.predict(X_test)
    mape_test = mape(y_test, test_pred)
    return(mape_test)
```

Create a generalized function to calculate the R-Squared and Adjusted R- Squared

```python
def get_score(model):
    r_sq = model.score(X_train, y_train)
    n = X_train.shape[0]
    k = X_train.shape[1]
    r_sq_adj = 1 - ((1-r_sq)*(n-1)/(n-k-1))
    return ([r_sq, r_sq_adj])
```

Create a generalized function to create a dataframe containing the scores from all the models

```python
score_card = pd.DataFrame(columns=['Model_Name', 'Alpha (Wherever Required)', 'l1-ratio', 'R-Squared',
                                       'Adj. R-Squared', 'Test_RMSE', 'Test_MAPE'])

def update_score_card(algorithm_name, model, alpha = '-', l1_ratio = '-'):
    
    
    global score_card

    
    score_card = score_card.append({'Model_Name': algorithm_name,
                       'Alpha (Wherever Required)': alpha, 
                       'l1-ratio': l1_ratio, 
                       'Test_MAPE': get_test_mape(model), 
                       'Test_RMSE': get_test_rmse(model), 
                       'R-Squared': get_score(model)[0], 
                       'Adj. R-Squared': get_score(model)[1]}, ignore_index = True)
```

# Now Creating generalized functions to calculate various metrics for models:

Create a generalized function to calculate the RMSE for train and test set.

```python
def get_train_rmse(model):
    train_pred = model.predict(New_X_train)
    mse_train = mean_squared_error(y_train, train_pred)
    rmse_train = round(np.sqrt(mse_train), 4)
    return(rmse_train)
```

```python
def get_test_rmse(model):
    test_pred = model.predict(New_X_test)
    mse_test = mean_squared_error(y_test, test_pred)
    rmse_test = round(np.sqrt(mse_test), 4)
    return(rmse_test)
```

Create a generalized function to calculate the MAPE for test set.

```python
def mape(actual, predicted):
    return (np.mean(np.abs((actual - predicted) / actual)) * 100)

def get_test_mape(model):
    test_pred = model.predict(New_X_test)
    mape_test = mape(y_test, test_pred)
    return(mape_test)
```

Create a generalized function to calculate the R-Squared and Adjusted R- Squared

```python
def get_score(model):
    r_sq = model.score(New_X_train, y_train)
    n = X_train.shape[0]
    k = X_train.shape[1]
    r_sq_adj = 1 - ((1-r_sq)*(n-1)/(n-k-1))
    return ([r_sq, r_sq_adj])
```

Create a generalized function to create a dataframe containing the scores from all the models

```python
score_card = pd.DataFrame(columns=['Model_Name', 'Alpha (Wherever Required)', 'l1-ratio', 'R-Squared',
                                       'Adj. R-Squared', 'Test_RMSE', 'Test_MAPE'])

def update_score_card(algorithm_name, model, alpha = '-', l1_ratio = '-'):
    
    
    global score_card

    
    score_card = score_card.append({'Model_Name': algorithm_name,
                       'Alpha (Wherever Required)': alpha, 
                       'l1-ratio': l1_ratio, 
                       'Test_MAPE': get_test_mape(model), 
                       'Test_RMSE': get_test_rmse(model), 
                       'R-Squared': get_score(model)[0], 
                       'Adj. R-Squared': get_score(model)[1]}, ignore_index = True)
```
